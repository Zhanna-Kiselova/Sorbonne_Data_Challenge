Le 30 mars 2023 au Grand Salon de la Sorbonne, un évènement a eu lieu sur le thème de l'économie d'énergie et le développement durable, organisé par l'Université Paris 1 Panthéon-Sorbonne, Engie et DataScientest. 
L'objectif de ce data challenge était de trouver le meilleur modèle de prédiction qui estimerait l'optimisation de production d'énergie par les éoliennes dans le département de la Meuse pour éviter la surproduction de l'énergie qui, à son retour, engendrait le coût de stockage et donc l'émission de CO2 mais également éviter la sousproduction qui nécessiterait de faire appel aux autres moyens d'énergies très polluants (le charbon, le pétrole, le gaz, etc). 
Nous avons reçu un dataset divisé en 3 fichiers avec les données sur lesquelles on devait entrainer nos modèles, les données sur lesquelles on devait tester nos modèles ainsi que les 
informations avec les données statiques.


```python
# Pour commencer, nous avons importé les libraries nécessaires. Aussi, nous créons les variables pour visualiser les données dans les dataframes. Nous analysons les tableaux.  
import pandas as pd
train = pd.read_csv('train_set.csv', sep=',')
test = pd.read_csv('test_set.csv', sep=',')
infos =pd.read_csv('static_desc.csv',sep=',')
train.head()
```

```python
# Nous vérifions la présence de valeurs manquantes dans le tableau TRAIN qui devront être soit remplacées par les valeurs moyennes soit supprimées
train.isna().sum(axis=0)
```

```python
# Mais d'abord, nous transformons ces valeurs manquantes de type string en NAN de la librarie numpy pour pouvoir leur attribuer la valeur moyenne numérique 
# et nous changeons également le type de tous nos variables pour float pour uniformiser sauf la 1ère colonne/target qui est la colonne catégorielle et la dernière qui est la colonne avec la date.   
import numpy as np 
train['Pitch_angle'] = train['Pitch_angle'].replace('Error', np.nan).astype('float')
train['Generator_converter_speed'] = train['Generator_converter_speed'].replace('Error', np.nan).astype('float')
train['Generator_temp_2'] = train['Generator_temp_2'].replace('Error', np.nan).astype('float')
train['Nacelle_angle'] = train['Nacelle_angle'].replace('Error', np.nan).astype('float')
train['Vanne_position_1'] = train['Vanne_position_1'].replace('Error', np.nan).astype('float')
train['Grid_frequency'] = train['Grid_frequency'].replace('Error', np.nan).astype('float')
train['Grid_voltage'] = train['Grid_voltage'].replace('Error', np.nan).astype('float')
train['Rotor_temp'] = train['Rotor_temp'].replace('Error', np.nan).astype('float')


print(train.info())
```

```python
# On remplace les valeurs manquantes numériques par les valeurs moyennes et les valeurs manquantes de type string pour la 1ère et la dernière colonnes par la valeur la plus utilisée 
# Nous constatons que les valeurs manquantes pour les deux colonnes "Vanne_position_1" & "Vanne_position_2" sont toujours là. On est obligées de les supprimer dans ce cas-là.  
num_cols=["Pitch_angle", "Hub_temperature", "Generator_converter_speed", "Generator_temp_1","Generator_temp_2", "Nacelle_angle", "Wind_direction", "Vanne_position_1", "Vanne_position_2", "Outdoor_temp", "Outdoor_temp_min", "Outdoor_temp_max", "Grid_frequency", "Grid_voltage", "Rotor_temp"]
for num in num_cols:
    train[num]=train[num].fillna(train[num].mean())
string_cols=["Wind_turbine_name", "date"]
for string in string_cols: 
    train[string]=train[string].fillna(train[string].mode()[0])

train.isna().sum(axis=0)
```

```python
train=train.drop(['Vanne_position_1'], axis=1)
train=train.drop(['Vanne_position_2'], axis=1)
train.head()
```

```python
# On supprime la colonne avec la date également 
train=train.drop(["date"], axis=1)
train.head()
```

```python
# On la remplace par la valeur catégorielle la plus utilisée 
infos["Region"].fillna("Grand Est", inplace = True)
infos.info()
```

```python
# On supprime les colonnes qui ne sont pas pertinentes 
infos = infos.drop(['Region','Department','Wind_turbine_long_name'], axis =1) 
infos.head()
```

```python
# Il y a 6 uniques noms de turbines 
infos['Wind_turbine_name'].unique()
```

```python
# On remplace les valeurs catégorielles en numériques 
train['Wind_turbine_name']=train["Wind_turbine_name"].replace(['R80736', 'R80721', 'R80711', 'R80790', 'R80789', 'R80925'], [1,2,3,4,5,6])
train.head()
```

```python
# On remplace les valeurs catégorielles avec les valeurs numériques et on divise la colonne GPS qui contient deux valeurs en deux colonnes  
infos['Wind_turbine_name']=infos["Wind_turbine_name"].replace(['R80736', 'R80721', 'R80711', 'R80790', 'R80789', 'R80925'], [1,2,3,4,5,6])
infos['Manufacturer']=infos["Manufacturer"].replace(['SENVION', 'EOLYS', 'NENUPHAR', 'VERGNET'], [1,2,3,4,])
infos['Model']=infos["Model"].replace(['MM82', 'EL80', 'NVP', 'MPC-V', 'NVP-2', 'MR-54'],[1,2,3,4,5,6] )
infos["GPS1"]=infos["GPS"].apply(lambda x: float(x.split(',') [0]))
infos["GPS2"]=infos["GPS"].apply(lambda x: float(x.split(',') [1]))
infos=infos.drop(["GPS"], axis=1)
infos.head()
```

```python
# On merge la colonne TRAIN et INFOS 
df_all = pd.merge(train, infos, on=['Wind_turbine_name'])
df_all
```

```python
# On importe le modèle de Regression Linéaire et le module train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X=df_all.drop('Power', axis =1) # X contiendra toutes les colonnes avec les valeurs sur lesquelles on va effectuer l'entrainement et le test 
y=df_all['Power'] # y va contenir les valeurs de notre colonne TARGET = POWER dont les valeurs devront être prédites 

X_norm= 2*(X-X.min())/(X.max()-X.min())-1 # On normalise les valeurs de X pour uniformiser et standardiser 

X_train, X_test, y_train, y_test = train_test_split(X_norm, y, test_size=0.3, random_state = 42) # avec 30% de jeu de données prises pour tester et 70% pour entrainer 
```

```python
print(X_train.info())
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_norm, y, test_size=0.3, random_state = 10)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()  # Instanciation du modèle

reg.fit(X_train, y_train)  # Ajustement sur les données d'entrainement

reg.score(X_train, y_train)  # Score sur les données test
```

```python
y_pred_linreg = reg.predict(X_test) # Prédiction de X sur le jeu de données test 

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
mse_test=mean_squared_error(y_test,y_pred_linreg)
print('MSE est :', mse_test)
mae_test=mean_absolute_error(y_test,y_pred_linreg)
print('MAE est :', mae_test)
```

```python
# On importe le modèle de Regression Lasso 
from sklearn.linear_model import Lasso 

lasso = Lasso()
lasso.fit(X_train, y_train)

lasso.score(X_train, y_train)
```

```python
y_pred_lasso = lasso.predict(X_test)

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
mse_test_lasso=mean_squared_error(y_test, y_pred_lasso)
print('MSE est :', mse_test_lasso)
mae_test_lasso=mean_absolute_error(y_test,y_pred_lasso)
print('MAE est :', mae_test_lasso)
```

```python
# On importe le modèle de Regression RandomForest qui nous montre le meilleur score sur le jeu de train 
from sklearn.ensemble import RandomForestRegressor 

random = RandomForestRegressor(random_state=100)
random.fit(X_train, y_train)
random.score(X_train, y_train)
```

```python
# Néanmoins l'erreur sur le jeu de test pour la prédiction y est assez élevée. Le modèle Random reste quand même le plus performant parmi les 3 testés. 
# Il sera judicieux de tester d'autres modèles et/ou ajouter les paramètres en plus pour trouver la meilleure performance.   
y_pred_random = random.predict(X_test)

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
mse_test_random=mean_squared_error(y_test, y_pred_random)
print('MSE est :', mse_test_random)
mae_test_random=mean_absolute_error(y_test,y_pred_random)
print('MAE est :', mae_test_random)
```

```python
from sklearn.model_selection import GridSearchCV
rf = RandomForestRegressor()
param_rf = {'n_estimators' : (150, 200)}

grid_rf = GridSearchCV(rf, param_rf, cv=3)
grid_rf.fit(X_train, y_train)
print("Meilleurs hyperparamètres RF : ", grid_rf.best_params_)
print("Score : ", grid_rf.best_score_)
```
